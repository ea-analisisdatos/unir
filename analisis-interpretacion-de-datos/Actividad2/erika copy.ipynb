Para hacer referencia a las regiones de los países con fines estadísticos, la UE ha desarrollado una clasificación conocida como NUTS (Nomenclatura de unidades territoriales estadísticas).<br>

NUTS divide cada país de la UE en tres niveles:

- NUTS 1: grandes regiones socioeconómicas<b
- NUTS 2: regiones básicas (para las políticas regionales)
- NUTS 3: pequeñas regiones (para diagnósticos específicos)
- La clasificación   NUTS 2024 es válida a partir del 1 de enero de 2024. En él se enumeran 92 regiones de nivel NUTS 1, 244 regiones de nivel NUTS 2 y 1 165 regiones de nivel NUTS 3.

NUTS se utiliza para:

- recogida, desarrollo y armonización de las estadísticas regionales europeas
- realización de análisis socioeconómicos de las regiones
- elaboración de las políticas regionales de la UE
- las regiones que pueden optar a la ayuda de la política de cohesión se han definido en el nivel NUTS 2
- el informe sobre la cohesión se ha elaborado hasta ahora principalmente a nivel NUTS 2

```python
import pandas as pd
import requests
from io import BytesIO

# URL del archivo Excel
url = "https://ec.europa.eu/eurostat/documents/345175/629341/NUTS2021-NUTS2024.xlsx/2b35915f-9c14-6841-8197-353408c4522d?t=1717505289640"

# Descargar el archivo Excel
respuesta = requests.get(url)
respuesta.raise_for_status()

# Leer todas las hojas del archivo en un diccionario de DataFrames
hojas = pd.read_excel(BytesIO(respuesta.content), sheet_name=None)

# Cargar y renombrar todos los DataFrames
dataframes = {}
for nombre_hoja, df in hojas.items():
    # Crear nombre válido para variable: sin espacios, en minúsculas, con guiones bajos
    nombre_variable = "df_" + nombre_hoja.strip().lower().replace(" ", "_").replace("-", "_")
    globals()[nombre_variable] = df  # Crea la variable con ese nombre
    dataframes[nombre_variable] = df  # También lo guarda en el diccionario si quieres mantener acceso agrupado
    print(f"{nombre_variable} cargado.")

# Ejemplo de acceso:
# df_statistical_regions.head()
```
